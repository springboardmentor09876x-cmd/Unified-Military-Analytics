```python
# 1. Cleaning the Text
def clean_text(value):
    if pd.isna(value):
        return value
    value = str(value)
    # Remove commas
    value = value.replace(",", "")
    # Remove symbols and units
    remove_items = ["%", "+", "$", "sq km", "km", "bbl", "Cu.M", "mt", "USD"]
    for item in remove_items:
        value = value.replace(item, "")
    return value.strip()
```

```python
text_cols = df.select_dtypes(include=["object", "string"]).columns
for col in text_cols:
    df[col] = df[col].apply(clean_text)
print(df)
```

```python
#2. Convert metrics to numeric formats
text_columns = ["country", "continent", "region", "nato_member"]
for col in df.columns:
    if col not in text_columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
df.head()
```

```python
# 3. Standardize column names
df.rename(columns={
    "total_military_manpower": "military_manpower",
    "total_military_aircraft": "total_aircraft",
    "total_military_helicopters": "total_helicopters",
    "total_naval_fleet_tonnage_mt": "naval_fleet_tonnage",
    "total_land_area_sq_km": "land_area_sq_km",
    "total_serviceable_airports": "serviceable_airports",
    "total_merchant_marine_fleet": "merchant_marine_fleet",
    "gdp_usd": "gdp"
}, inplace=True)
```
