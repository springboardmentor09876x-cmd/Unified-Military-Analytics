```python
import pandas as pd
import numpy as np
import re
```

```python
# Load merged dataset
df = pd.read_csv("merged_military_dataset.csv")

print("Dataset Loaded Successfully!")
print("Original Shape:", df.shape)

df.head()
```

```python
# Standardize column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_", regex=False)
    .str.replace("-", "_", regex=False)
    .str.replace("/", "_", regex=False)
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)

print("Column names standardized successfully.")

df.head()
```

```python
# Remove extra spaces from text columns

text_columns = df.select_dtypes(include=["object"]).columns

for col in text_columns:
    df[col] = df[col].astype(str).str.strip()

print("Extra spaces removed.")
```

```python
# Function to clean numeric values

def clean_numeric(value):

    if pd.isna(value):
        return np.nan

    value = str(value)

    # Remove commas
    value = value.replace(",", "")

    # Remove percentage sign
    value = value.replace("%", "")

    # Remove plus sign
    value = value.replace("+", "")

    # Remove dollar symbol
    value = value.replace("$", "")

    # Remove brackets
    value = value.replace("(", "")
    value = value.replace(")", "")

    # Remove units
    value = re.sub(r"km²|sq km|km|mt|bbl|usd|cum", "", value, flags=re.IGNORECASE)

    # Remove unwanted characters
    value = re.sub(r"[^\d\.\-]", "", value)

    if value == "":
        return np.nan

    return value

# Apply cleaning
for col in text_columns:

    if col not in [
        "country",
        "metric",
        "source_url",
        "region",
        "continent",
        "alliance",
        "major_ports_and_terminals"
    ]:

        df[col] = df[col].apply(clean_numeric)

print("Special symbols removed successfully.")
```

```python
# Convert columns to numeric

for col in df.columns:

    if col not in [
        "country",
        "metric",
        "source_url",
        "region",
        "continent",
        "alliance",
        "major_ports_and_terminals"
    ]:

        df[col] = pd.to_numeric(df[col], errors="coerce")

print("Numeric conversion completed.")

df.info()
```

```python
# Fill missing numeric values with median

numeric_columns = df.select_dtypes(include=["number"]).columns

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].median())

# Fill missing categorical values with mode

categorical_columns = df.select_dtypes(include=["object"]).columns

for col in categorical_columns:

    if not df[col].mode().empty:
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna("Unknown")

print("Missing values handled successfully.")
```

```python
# Remove completely empty columns

missing_percentage = (df.isnull().sum() / len(df)) * 100

columns_to_drop = missing_percentage[missing_percentage == 100].index.tolist()

if columns_to_drop:
    print("Dropping Columns:", columns_to_drop)
    df.drop(columns=columns_to_drop, inplace=True)
else:
    print("No columns with 100% missing values found.")
```

```python
before = len(df)

df.drop_duplicates(inplace=True)

after = len(df)

print("Duplicate Rows Removed:", before - after)
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100

print("Missing Percentage Per Column")

print(missing_percentage)

print("\nMaximum Missing Percentage:")

print(round(missing_percentage.max(), 2), "%")
```

```python
output_file = "final_military_cleaned.csv"

df.to_csv(output_file, index=False)

print(f"Dataset saved successfully as '{output_file}'")
```
