```python
import pandas as pd
import numpy as np
import re

# Load dataset
df = pd.read_csv("military_data_raw.csv")

# -----------------------------
# Column Names
# -----------------------------
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
)

# -----------------------------
# Remove Duplicate Rows
# -----------------------------
df = df.drop_duplicates()

# -----------------------------
# Clean Country Names
# -----------------------------
df["country"] = (
    df["country"]
    .astype(str)
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)

# -----------------------------
# Clean All Other Columns
# -----------------------------
for col in df.columns:

    if col == "country":
        continue

    # Convert to string
    df[col] = df[col].astype(str)

    # Remove tabs/newlines
    df[col] = df[col].str.replace(r"[\t\r\n]", "", regex=True)

    # Remove extra spaces
    df[col] = df[col].str.replace(r"\s+", " ", regex=True)

    # Remove commas
    df[col] = df[col].str.replace(",", "", regex=False)

    # Remove $
    df[col] = df[col].str.replace("$", "", regex=False)

    # Remove %
    df[col] = df[col].str.replace("%", "", regex=False)

    # Keep only numbers, decimal point and minus sign
    df[col] = df[col].str.replace(r"[^0-9.\-]", "", regex=True)

    # Empty -> NaN
    df[col] = df[col].replace("", np.nan)

    # Convert to numeric
    df[col] = pd.to_numeric(df[col], errors="coerce")

# -----------------------------
# Remove duplicate countries
# -----------------------------
df = df.drop_duplicates(subset="country")

# -----------------------------
# Fill Missing Values
# -----------------------------
for col in df.columns:
    if col != "country":
        df[col] = df[col].fillna(0)

# -----------------------------
# Save
# -----------------------------
df.to_csv("military_data_cleaned.csv", index=False)

print("Cleaning Completed Successfully")
print(df.info())
print(df.head())
```
