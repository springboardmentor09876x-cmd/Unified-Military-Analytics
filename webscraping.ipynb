```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.globalfirepower.com/aircraft-total.php"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

all_data = []

countries = soup.find_all("div", class_="recordsetContainer")

for country in countries:

    rank = country.find("div", class_="rankNumContainer").get_text(strip=True)

    country_name = country.find("div", class_="longFormName").get_text(strip=True)

    aircraft = country.find("div", class_="valueContainer").get_text(strip=True)

    aircraft = aircraft.replace(",", "")

    all_data.append({
        "Rank": rank,
        "Country": country_name,
        "Total Aircraft": aircraft
    })

df = pd.DataFrame(all_data)

df.to_csv("aircraft_total.csv", index=False)

print(df.head())
print(f"\nTotal Countries : {len(df)}")
print("CSV Created Successfully")
```
