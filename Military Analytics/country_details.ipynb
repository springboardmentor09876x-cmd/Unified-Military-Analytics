```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
# Read countries list
df = pd.read_csv("data/countries.csv")

print("Total Countries:", len(df))
df.head()
```

```python
def scrape_country_details(url):
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")

    data = {}

    # Country Name
    try:
        data["Country"] = soup.find("h1").text.strip()
    except:
        data["Country"] = ""

    # Power Index
    try:
        power = soup.find("div", class_="contentboxheader")
        data["Power_Index"] = power.text.strip()
    except:
        data["Power_Index"] = ""

    # Store all specifications
    boxes = soup.find_all("div", class_="contentbox")

    for box in boxes:

        title = box.find("span")

        value = box.find("div", class_="textLarge")

        if title and value:

            key = title.text.strip()

            val = value.text.strip()

            data[key] = val

    return data
```

```python
all_data = []

for index, row in df.iterrows():

    print(f"{index+1}. {row['Country']}")

    details = scrape_country_details(row["Detail_URL"])

    details["Rank"] = row["Rank"]

    details["Country"] = row["Country"]

    details["Power_Index"] = row["Power_Index"]

    all_data.append(details)

    time.sleep(1)
```

```python
details_df = pd.DataFrame(all_data)

details_df.head()
```

```python
print(details_df.columns)
```

```python
details_df.to_csv("data/country_details.csv", index=False)

print("country_details.csv saved successfully")
```
