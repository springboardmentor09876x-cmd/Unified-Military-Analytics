```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.globalfirepower.com/aircraft-total.php"

headers = {
    "User-Agent": "Mozilla/5.0"
}

page = requests.get(url, headers=headers)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup).prettify()
```

```python
divs = soup.find_all('div', class_='rankNumContainer')

count = 0
for div in divs:
    count += 1
    print(div.get_text(strip=True))

print(count)
```

```python
divs = soup.find_all('div', class_='longFormName')

count = 0
for div in divs:
    count += 1
    print(div.get_text(strip=True))

print(count)
```

```python
divs = soup.find_all('div', class_='valueContainer')

count = 0
for div in divs:
    count += 1
    print(div.get_text(strip=True))

print(count)
```

```python
import pandas as pd
```

```python
ranks = []
for div in soup.find_all('div', class_='rankNumContainer'):
    ranks.append(div.get_text(strip=True))

names = []
for div in soup.find_all('div', class_='longFormName'):
    names.append(div.get_text(strip=True))

values = []
for div in soup.find_all('div', class_='valueContainer'):
    values.append(div.get_text(strip=True))

df = pd.DataFrame({
    "Rank": ranks,
    "Country": names,
    "Aircraft Fleet": values
})
```
