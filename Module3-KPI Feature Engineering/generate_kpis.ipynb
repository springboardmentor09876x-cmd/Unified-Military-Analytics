```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

print("Libraries imported successfully!")
```

```python
df = pd.read_csv("military_cleaned.csv")

print("Dataset loaded successfully!")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.head()
```

```python
print("Column Names:\n")

for col in df.columns:
    print(col)
```

```python
df["power_index_rank"] = (
    df["power_index"]
      .rank(ascending=True, method="min")
      .astype(int)
)

print("✅ Power Index Rank created successfully!")

df[["country", "power_index", "power_index_rank"]].head()
```

```python
asset_columns = [
    "total_military_aircraft",
    "tanks",
    "armored_fighting_vehicles",
    "self_propelled_artillery",
    "towed_artillery",
    "rocket_projectors",
    "total_naval_fleet"
]

df["total_assets"] = df[asset_columns].fillna(0).sum(axis=1)

print("✅ Total Assets calculated successfully!")

df[["country", "total_assets"]].head()
```

```python
df["gdp_rank"] = (
    df["gdp"]
      .rank(ascending=False, method="min")
      .astype(int)
)

print("✅ GDP Rank created successfully!")

df[["country", "gdp", "gdp_rank"]].head()
```

```python
df["power_index_rank_gap"] = (
    df["gdp_rank"] -
    df["power_index_rank"]
)

print("✅ Power Index Rank Gap calculated successfully!")

df[[
    "country",
    "power_index_rank",
    "gdp_rank",
    "power_index_rank_gap"
]].head()
```

```python
df["assets_per_capita"] = (
    df["total_assets"] /
    df["total_population"]
)

print("✅ Assets per Capita calculated successfully!")

df[[
    "country",
    "total_assets",
    "total_population",
    "assets_per_capita"
]].head()
```

```python
df["budget_to_gdp_ratio"] = (
    df["defense_budget_usd"] /
    df["gdp"]
)

print("✅ Budget-to-GDP Ratio calculated successfully!")

df[[
    "country",
    "defense_budget_usd",
    "gdp",
    "budget_to_gdp_ratio"
]].head()
```

```python
df["nato_flag"] = np.where(
    df["alliance"].astype(str).str.contains("NATO", case=False, na=False),
    1,
    0
)

print("✅ NATO Flag created successfully!")

df[["country", "alliance", "nato_flag"]].head()
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)

df.sort_values(
    by="country",
    key=lambda x: x.str.lower(),
    inplace=True
)

df.reset_index(drop=True, inplace=True)

print("✅ Dataset cleaned and sorted successfully!")

df.head()
```

```python
id_columns = [
    "country",
    "continent",
    "region",
    "alliance",
    "nato_flag"
]

kpi_columns = [
    "power_index",
    "power_index_rank",
    "gdp_rank",
    "power_index_rank_gap",
    "total_assets",
    "assets_per_capita",
    "budget_to_gdp_ratio"
]

long_df = df[id_columns + kpi_columns].melt(
    id_vars=id_columns,
    var_name="kpi_name",
    value_name="kpi_value"
)

print("✅ Long Format created successfully!")
print("Rows:", long_df.shape[0])

long_df.head()
```

```python
definitions = pd.DataFrame({
    "KPI": [
        "Power Index Rank Gap",
        "Assets per Capita",
        "Budget-to-GDP Ratio",
        "Power Index Rank",
        "GDP Rank",
        "NATO Flag"
    ],
    "Formula": [
        "GDP Rank - Power Index Rank",
        "Total Assets / Total Population",
        "Defense Budget USD / GDP",
        "Rank based on Power Index",
        "Rank based on GDP",
        "1 = NATO Member, 0 = Non-NATO"
    ]
})

print("✅ KPI Definitions created successfully!")

definitions
```

```python
with pd.ExcelWriter(
    "military_final.xlsx",
    engine="openpyxl"
) as writer:

    df.to_excel(
        writer,
        sheet_name="Wide_Format",
        index=False
    )

    long_df.to_excel(
        writer,
        sheet_name="Long_Format",
        index=False
    )

    definitions.to_excel(
        writer,
        sheet_name="KPI_Definitions",
        index=False
    )

print("✅ military_final.xlsx created successfully!")
```

```python
wide = pd.read_excel(
    "military_final.xlsx",
    sheet_name="Wide_Format"
)

long = pd.read_excel(
    "military_final.xlsx",
    sheet_name="Long_Format"
)

print("=" * 50)
print("MODULE 3 VERIFICATION")
print("=" * 50)

print("Wide Format Shape :", wide.shape)
print("Long Format Shape :", long.shape)

required_kpis = [
    "power_index_rank_gap",
    "assets_per_capita",
    "budget_to_gdp_ratio",
    "nato_flag"
]

print("\nRequired KPI Columns:")

for col in required_kpis:
    print(f"{col}: {'Present' if col in wide.columns else 'Missing'}")

print("\n✅ Module 3 completed successfully!")
```
