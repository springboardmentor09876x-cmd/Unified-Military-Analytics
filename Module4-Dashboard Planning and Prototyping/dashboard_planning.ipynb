```python
df = pd.read_excel(
    "military_final.xlsx",
    sheet_name="Wide_Format"
)

print("Dataset Loaded Successfully")

print("Rows :",df.shape[0])

print("Columns :",df.shape[1])

df.head()
```

```python
countries = len(df)
aircraft = int(df["total_military_aircraft"].sum())
budget = df["defense_budget_usd"].sum() / 1e9
power = round(df["power_index"].mean(), 4)

display(HTML(f"""

<div style="display:flex;
gap:25px;
justify-content:center;
margin-top:20px;">

<div class="card"
title="Total number of countries available in the dataset">

<h3>🌍 Countries</h3>

<h1>{countries}</h1>

</div>

<div class="card"
title="Total military aircraft across all countries">

<h3>✈ Aircraft</h3>

<h1>{aircraft:,}</h1>

</div>

<div class="card"
title="Combined defense budget of all countries (in Billion USD)">

<h3>💰 Budget</h3>

<h1>{budget:.2f} B</h1>

</div>

<div class="card"
title="Average Global Firepower Index (lower is stronger)">

<h3>⚔ Avg Power</h3>

<h1>{power}</h1>

</div>

</div>

"""))
```

```python
import plotly.express as px

top10 = df.sort_values(
    by="power_index",
    ascending=True
).head(10)

fig = px.bar(
    top10,
    x="country",
    y="power_index",
    color="power_index",          # Automatic gradient
    text="power_index",
    color_continuous_scale="Brwnyl",   # Brown → Yellow gradient
    title="🏆 Top 10 Countries by Power Index"
)

fig.update_traces(
    texttemplate="%{text:.3f}",
    textposition="outside",
    hovertemplate="<b>%{x}</b><br>Power Index: %{y:.3f}<extra></extra>"
)

fig.update_layout(
    paper_bgcolor="#F7F2E8",
    plot_bgcolor="white",
    height=550,
    title_x=0.5,
    xaxis_title="Country",
    yaxis_title="Power Index",
    coloraxis_showscale=False
)

fig.show()
```

```python
top_budget = df.sort_values(
    by="defense_budget_usd",
    ascending=False
).head(10)

fig = px.bar(
    top_budget,
    x="country",
    y="defense_budget_usd",
    color="defense_budget_usd",
    text="defense_budget_usd",
    color_continuous_scale="YlOrBr",
    title="💰 Top 10 Defense Budgets"
)

fig.update_traces(
    texttemplate="%{text:.2s}",
    textposition="outside",
    hovertemplate="<b>%{x}</b><br>Defense Budget: $%{y:,.0f}<extra></extra>"
)

fig.update_layout(
    paper_bgcolor="#F7F2E8",
    plot_bgcolor="white",
    title_x=0.5,
    title_font_size=22,
    height=550,
    xaxis_title="Country",
    yaxis_title="Defense Budget (USD)",
    coloraxis_showscale=False
)

fig.show()
```

```python
continent_assets = (
    df.groupby("continent")["total_assets"]
      .sum()
      .reset_index()
      .sort_values("total_assets", ascending=False)
)


colors = [
    "#0B1F3A",  # Navy Blue
    "#1F4E79",  # Dark Blue
    "#2E75B6",  # Steel Blue
    "#5B9BD5",  # Medium Blue
    "#9DC3E6",  # Light Blue
    "#D9EAF7",  # Pale Blue
    "#F4F8FC"   # Very Light Blue
]
fig = px.pie(
    continent_assets,
    names="continent",
    values="total_assets",
    hole=0.55,
    title="🌍 Military Assets by Continent",
    color_discrete_sequence=colors
)

fig.update_traces(
    textposition="inside",
    textinfo="percent+label",
    hovertemplate="<b>%{label}</b><br>Total Assets: %{value:,}<br>%{percent}<extra></extra>",
    marker=dict(line=dict(color="white", width=2))
)

fig.update_layout(
    paper_bgcolor="#F7F2E8",
    plot_bgcolor="#F7F2E8",
    title_x=0.5,
    title_font_size=22,
    height=600,
    legend_title="Continent"
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="total_military_aircraft",
    y="tanks",
    size="defense_budget_usd",
    color="continent",
    hover_name="country",
    title="✈ Aircraft vs Tanks by Country",
    color_discrete_sequence=px.colors.qualitative.Bold
)

fig.update_layout(
    paper_bgcolor="#F7F2E8",
    plot_bgcolor="white",
    title_x=0.5,
    title_font_size=22,
    height=600,
    xaxis_title="Total Military Aircraft",
    yaxis_title="Number of Tanks"
)

fig.update_traces(
    marker=dict(
        line=dict(width=1, color="white")
    )
)

fig.show()
```

```python
region_data = (
    df.groupby("region")["active_personnel"]
      .sum()
      .reset_index()
      .sort_values("active_personnel", ascending=False)
)

fig = px.bar(
    region_data,
    x="region",
    y="active_personnel",
    color="active_personnel",
    text="active_personnel",
    color_continuous_scale="YlOrBr",
    title="👥 Active Personnel by Region"
)

fig.update_traces(
    texttemplate="%{text:.2s}",
    textposition="outside",
    hovertemplate="<b>%{x}</b><br>Active Personnel: %{y:,}<extra></extra>"
)

fig.update_layout(
    paper_bgcolor="#F7F2E8",
    plot_bgcolor="white",
    title_x=0.5,
    title_font_size=22,
    height=550,
    xaxis_title="Region",
    yaxis_title="Active Personnel",
    coloraxis_showscale=False
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="gdp",
    y="defense_budget_usd",
    size="total_assets",
    color="power_index",
    hover_name="country",
    color_continuous_scale="YlOrBr",
    title="💰 Defense Budget vs GDP",
    log_x=True,
    log_y=True
)

fig.update_layout(
    paper_bgcolor="#F7F2E8",
    plot_bgcolor="white",
    title_x=0.5,
    height=600,
    xaxis_title="GDP (Log Scale)",
    yaxis_title="Defense Budget (Log Scale)"
)

fig.show()
```

```python
fig = px.choropleth(
    df,
    locations="country",
    locationmode="country names",
    color="power_index",
    hover_name="country",
    color_continuous_scale="YlOrBr",
    title="🌍 Global Military Power Index"
)

fig.update_layout(
    paper_bgcolor="#F7F2E8",
    title_x=0.5,
    height=600
)

fig.show()
```

```python
top10 = df.sort_values(
    "active_personnel",
    ascending=False
).head(10)

fig = px.bar(
    top10,
    x="country",
    y="active_personnel",
    color="active_personnel",
    text="active_personnel",
    color_continuous_scale="YlOrBr",
    title="👥 Top 10 Countries by Active Personnel"
)

fig.update_traces(
    texttemplate="%{text:.2s}",
    textposition="outside"
)

fig.update_layout(
    paper_bgcolor="#F7F2E8",
    plot_bgcolor="white",
    title_x=0.5,
    height=550,
    coloraxis_showscale=False
)

fig.show()
```
